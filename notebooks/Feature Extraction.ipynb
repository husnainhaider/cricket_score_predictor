```python
import pandas as pd
import pickle
import numpy as np
import os
```

```python
model_path = os.path.join('..', 'model')
df = pickle.load(open(model_path+'\psl_dataset_2.pkl','rb'))
```

```python
df
#required features
# batting_team
# bowling team
# city
# current score
# balls left
# wickets left
# current run rate
# last five
```

```python
df.isnull().sum()
```

```python
df[df['city'].isnull()]['venue'].value_counts()
```

```python
cities = np.where(df['city'].isnull(),df['venue'].str.split().apply(lambda x:x[0]),df['city'])
```

```python
df['city'] = cities
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['venue'],inplace=True)
```

```python
df['city'].value_counts()
```

```python
#cities where a minimum of 5 innings have been played
eligible_cities = df['city'].value_counts()[df['city'].value_counts() > 600].index.tolist()
```

```python
eligible_cities
```

```python
df = df[df['city'].isin(eligible_cities)]
```

```python
df
```

```python
df['runs'] = pd.to_numeric(df['runs'], errors='coerce')
```

```python
#using cumulative sum to find the current score
df['current_score'] = df.groupby('match_id')['runs'].cumsum()
```

```python
df
```

```python
#splitting ball column to get over and ball no
df['over'] = df['ball'].apply(lambda x:str(x).split(".")[0])
df['ball_no'] = df['ball'].apply(lambda x:str(x).split(".")[1])
df
```

```python
df
```

```python
#balls bowled
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')
df
```

```python
#wickets left
df['player_dismissed'] = df['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
df['player_dismissed'] = df['player_dismissed'].astype('int')
df['player_dismissed'] = df.groupby('match_id')['player_dismissed'].cumsum()
df['wickets_left'] = 10 - df['player_dismissed']
```

```python
df
```

```python
df['crr'] = (df['current_score']*6)/df['balls_bowled']
```

```python
df
```

```python
# Group by match_id
groups = df.groupby('match_id')

# Get unique match IDs
match_ids = df['match_id'].unique()

# Initialize an empty list to store results
last_five = []

# Loop over each match ID
for id in match_ids:
    # Get the group for the current match_id
    group = groups.get_group(id)
    
    # Ensure that the 'runs' column is numeric
    group['runs'] = pd.to_numeric(group['runs'], errors='coerce')
    
    # Apply rolling window of 30 balls (5 overs) and sum the runs
    rolling_sums = group['runs'].rolling(window=30, min_periods=1).sum()
    
    # Extend the list with the rolling sums
    last_five.extend(rolling_sums.tolist())

# Output the result
last_five
```

```python
df['last_five'] = last_five
```

```python
df
```

```python
#total runs
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
```

```python
final_df=final_df[['batting_team','bowling_team','city','current_score','wickets_left','crr','last_five','runs_x']]
```

```python
final_df
```

```python
final_df.isnull().sum()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df
```

```python
#shuffling to avoid bias
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.head()
```

# Model Training

```python
X = final_df.drop(columns=['runs_x'])
y = final_df['runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city'])
], remainder='passthrough')
```

```python
#random forest
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',RandomForestRegressor(n_estimators=300, random_state=101))
])
```

```python
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
model_path = os.path.join('..', 'model')
pickle.dump(pipe,open(model_path+'\psl.pkl','wb'))
```
